```python
df= pd.read_csv("/home/jovyan/SnowEx-Map-Tool/GM_Pits.csv")

df['PitID'] = df['PitID'].str.strip()
grouped.index = grouped.index.str.strip()

df.head()
```

```python
from datetime import datetime 
# Find some density pit measurements at the Boise site in december 2019.
df1 = LayerMeasurements.from_filter(
    site_name="Grand Mesa",
    date_less_equal=datetime(2020, 5, 31),
    date_greater_equal=datetime(2019, 12, 1),
    limit=100000
)

# Plot Example!
df1.plot()

# Show off the dataframe
df1

```

```python
grouped = df1.groupby('pit_id')['doi'].apply(list)
print(grouped.head())
```

```python
# Merge the aggregated DOIs with df
merged_df = df.merge(grouped, left_on='PitID', right_on='pit_id', how='left')
```

```python
merged_df.tail()
```

```python
import folium

# Create a map centered around the average latitude and longitude of your points
map_center = [df['Latitude (deg)'].mean(), df['Longitude (deg)'].mean()]
m = folium.Map(location=map_center, zoom_start=10)

# Add markers with popups
for _, row in merged_df.iterrows():
    popup_text = f"PitID: {row['PitID']}<br>DOIs:<br>{row['doi']}"
    folium.Marker(
        location=[row['Latitude (deg)'], row['Longitude (deg)']],
        popup=popup_text
    ).add_to(m)

# Save or display the map
m.save('snow_pit_map.html')
```
