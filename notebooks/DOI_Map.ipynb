```python
import pandas as pd
import numpy as np
```

```python
from snowexsql.api import PointMeasurements
site_name="Grand Mesa"
instrument= "pit ruler"

from datetime import datetime 
# Find all Grand Mesa pit ruler data
df2 = PointMeasurements.from_filter(
    site_name= site_name,
    instrument= instrument,
    limit=1000
)
print(df2[['doi', 'type', 'latitude', 'longitude']])
```

```python
import folium

# Create a map centered around the mean location
m = folium.Map(location=[df2['latitude'].mean(), df2['longitude'].mean()], zoom_start=12)

# Add the points to the map
for _, row in df2.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        icon=folium.Icon(color="green"),
        popup=f"Site: {row['doi']}",
    ).add_to(m)

# Save the map to an HTML file
m.save('/home/jovyan/SnowEx-Map-Tool/map.html')
```
